$E.$ Wyznacz i wyświetl macierz korelacji dla zbioru bez ostatniego atrybutu (2 pkt). Wskaż (1 pkt) lub wyznacz stosując odpowiednią formułę (3 pkt) liczbę atrybutów:
* silnie skorelowanych dodatnio (>=0.6)
* słabo skorelowanych dodatnio (>=0.2 ale <0.6)
* nieskorelowanych (>-0.2 i <0.2)
* słabo skorelowanych ujmenie (<=-0.2 ale >-0.6)
* silnie skorelowanych ujemnie (<= -0.6)


```python
#macierz korelacji
dane_korelacji = dane.drop(columns='klasa').corr()
dane_korelacji
```

```python
# miejsce na kod

# zamiast '-1' wstaw odpowiednie wartości lub formuły

sprawdzian['silnie dodatnio'] = (dane_korelacji[dane_korelacji >= 0.6].count().sum() - len(list(dane_korelacji)))/2 
# zamiast '-1' powinna być liczba atrybutów spełniających kryterium
sprawdzian['słabo dodatnio'] = (dane_korelacji[(dane_korelacji >= 0.2) & (dane_korelacji < 0.6)].count().sum() - len(list(dane_korelacji)))/2 
sprawdzian['brak korelacji'] = (dane_korelacji[(dane_korelacji > -0.2) & (dane_korelacji < 0.2)].count().sum() - len(list(dane_korelacji)))/2 
sprawdzian['słabo ujemnie'] = (dane_korelacji[(dane_korelacji <= -0.2) & (dane_korelacji > -0.6)].count().sum() - len(list(dane_korelacji)))/2 
sprawdzian['silnie ujemnie'] = (dane_korelacji[dane_korelacji <= -0.6].count().sum() - len(list(dane_korelacji)))/2 
zapisz(nazwa_pliku,sprawdzian) # zapis - tu nic nie zmieniaj

# zamiast '-1' wstaw odpowiednie wartości lub formuły
```

$L.$ Zastosuj na danych wybraną metodę grupowania (4 pkt). Wyniki grupowania porównaj z klasami na wykresach punktowych (2 pkt) oraz numerycznie przy pomocy macierzy pomyłek (2 pkt). Wyciągnij i opisz wnioski (3 pkt).

```python
# miejsce na kod - grupowanie
from sklearn.cluster import KMeans
df = dane.drop(columns='klasa').dropna()
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(df)
etykiety_klastrow = kmeans.fit_predict(df)
```

```python
# miejsce na kod - wykresy
plt.figure(figsize=(10,10))
plt.scatter(df.A, df.B, c=etykiety_klastrow)
```
